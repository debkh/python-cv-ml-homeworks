```python
# Loading image, converting color space to RGB and displaying it via pyplot
import cv2
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
sourceImagePath = '../../data/lesson_1/campbells.png'
img = cv2.imread(sourceImagePath)
if img is None:
    raise FileNotFoundError(f"Image not found at {sourceImagePath}")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# end of init and setup phase

plt.axis('off')
plt.imshow(img)
```

```python
# Pop Art collage via channel swapping
import cv2
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
sourceImagePath = '../../data/lesson_1/campbells.png'
img = cv2.imread(sourceImagePath)
if img is None:
    raise FileNotFoundError(f"Image not found at {sourceImagePath}")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# end of init and setup phase

red, green, blue = cv2.split(img)
imgRGB = cv2.merge([red, green, blue])
imgRBG = cv2.merge([red, blue, green])
imgGRB = cv2.merge([green, red, blue])
imgBGR = cv2.merge([blue, green, red])

imageLine1 = np.hstack((imgRGB, imgRBG))
imageLine2 = np.hstack((imgGRB, imgBGR))
resultImage = np.vstack((imageLine1, imageLine2))
plt.axis('off')
plt.imshow(resultImage)
```
